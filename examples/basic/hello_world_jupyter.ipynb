### Temel Pekiştirmeli Öğrenme Örneği: Stable Baselines3 ile CartPole

Bu örnekte, Stable Baselines3 kütüphanesini kullanarak klasik `CartPole-v1` ortamında bir PPO (Proximal Policy Optimization) ajanını eğiteceğiz. CartPole ortamının amacı, bir arabayı sağa veya sola hareket ettirerek direğin düşmesini engellemektir.

```python
# Gerekli kütüphaneleri yükleyin
!pip install gymnasium stable-baselines3
```

```python
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env

# Ortamı oluşturun
env_id = 'CartPole-v1'
vec_env = make_vec_env(env_id, n_envs=1)

print(f"Ortam oluşturuldu: {env_id}")
```

```python
# PPO modelini tanımlayın ve eğitin
# MlpPolicy (Çok Katmanlı Algılayıcı Politika) kullanıyoruz çünkü gözlem ve eylem uzayları süreklidir.
model = PPO('MlpPolicy', vec_env, verbose=1)
print("Model eğitiliyor...")
model.learn(total_timesteps=10000) # 10000 zaman adımında eğitim
print("Model eğitimi tamamlandı.")
```

```python
# Eğitilmiş modeli değerlendirin

obs = vec_env.reset()

num_episodes = 5
episode_rewards = []

print("Model değerlendiriliyor...")
for episode in range(num_episodes):
    done = False
    episode_reward = 0
    while not done:
        action, _states = model.predict(obs, deterministic=True)
        obs, reward, done, info = vec_env.step(action)
        episode_reward += reward[0]
    episode_rewards.append(episode_reward)
    obs = vec_env.reset() # Her bölüm sonunda ortamı sıfırla

print(f"\n{num_episodes} bölümdeki ortalama ödül: {sum(episode_rewards) / num_episodes:.2f}")

# Ortamı kapatın
vec_env.close()
```

1. Üst menüden **Düzenle (Edit) > Not Defteri Ayarları (Notebook Settings)** yolunu izleyin.
2. **Donanım Hızlandırıcı (Hardware Accelerator)** olarak **T4 GPU** seçin.
3. **Kaydet** butonuna basın.

GPU aktif edildikten sonra kuruluma devam edebilirsiniz.

### PROF DARVIS Sistem Mimarisi

PROF DARVIS, kullanıcının dijital bir 'İkinci Benlik' (Second Self) vizyonuna hizmet eden, yerel kaynakları (RAG) ve gelişmiş dil modellerini (LLM) harmanlayan hiyerarşik bir yapıya sahiptir.

#### 1. Veri Akışı ve Katmanlar
Kullanıcıdan gelen bir girdi (Metin veya Ses), sistem içerisinde şu sırayı takip eder:
*   **Arayüz Katmanı (Gradio):** Kullanıcı etkileşimini sağlar. Ses girdileri Whisper veya benzeri bir modelle metne dönüştürülür.
*   **Karar Verme Mekanizması:** Girdinin türüne göre (genel sohbet, teknik soru, kişisel not sorgusu) ilgili modüllere yönlendirme yapar.
*   **RAG Modülü (LangChain & FAISS):** Girdi vektörleştirilir ve FAISS indeksli yerel veri tabanında (PDF, Kod, Notlar) benzerlik araması yapılır. İlgili bağlam (context) çekilir.
*   **LLM Katmanı (llama.cpp):** Orijinal sorgu ve RAG'den gelen bağlam birleştirilerek LLM'e (Llama 3 vb.) sunulur ve nihai yanıt üretilir.

#### 2. İndeksleme ve Geri Çağırma (Retrieval)
*   **İndeksleme:** Yerel PDF'ler, kod dosyaları ve kişisel notlar parçalara (chunks) bölünür, gömme (embedding) modelleriyle vektörlere dönüştürülür ve FAISS üzerinde saklanır.
*   **Geri Çağırma:** Kullanıcı sorusuyla en yüksek kosinüs benzerliğine sahip doküman parçaları, LLM'in yanıtını desteklemek üzere 'Context' olarak enjekte edilir.

#### 3. Dış Erişim ve Bağlantı Şeması
Sistem Google Colab üzerinde çalışırken, yerel sunucu (localhost) **Ngrok** tünelleme hizmeti kullanılarak dış dünyaya açılır. Bu sayede mobil cihazlar veya farklı tarayıcılar üzerinden güvenli bir URL ile sisteme erişim sağlanır.

#### 4. Bellek ve Hafıza Yönetimi (İkinci Benlik Stratejisi)
Sistemin sürekliliğini sağlamak için **ConversationBufferMemory** veya veritabanı tabanlı bir geçmiş (history) yönetimi kullanılır. Geçmiş konuşmalar, kullanıcının tercihlerini ve önceki bağlamları hatırlamak üzere 'Kısa Süreli Hafıza' olarak saklanırken, RAG modülü 'Uzun Süreli Bilgi Deposu' görevini görür.
